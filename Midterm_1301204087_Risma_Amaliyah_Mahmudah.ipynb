```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/rismaamaliyah/Data-Visualization-/main/diabetes_prediction_dataset.csv')
data.head(20)
```

```python
data = data[data['Gender'] != 'Other']
data = data[data['Smoking'] != 'No Info']
df = data.sample(frac=0.3)
df.columns = list(map(str, data.columns))
```

```python
df_Gender = df.groupby('Gender', axis=0)['Diabetes'].sum()

plt.figure(figsize=(8, 8))
plt.pie(df_Gender, labels=df_Gender.index, autopct='%1.1f%%', startangle=140, colors=['#66b3ff', '#99ff99', '#ffcc99'])
plt.title('Persentase Diabetes Berdasarkan Gender')
plt.show()
```

Pie chart di atas menunjukkan persentase penderita diabetes berdasarkan jenis kelamin.

Dari pie chart tersebut, dapat dilihat bahwa:
*   Perempuan lebih banyak menderita diabetes daripada laki-laki.
*   Persentase penderita diabetes pada perempuan adalah 53,4%, sedangkan persentase penderita diabetes pada laki-laki adalah 46,6%.

Berdasarkan hasil tersebut, dapat disimpulkan bahwa perempuan memiliki risiko lebih tinggi untuk menderita diabetes daripada laki-laki.

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], bins=20, kde=True, color='pink')
plt.title('Distribusi Usia')
plt.xlabel('Usia')
plt.ylabel('Frekuensi')
plt.show()
```

Grafik tersebut menunjukkan distribusi usia responden.

Dari grafik tersebut, dapat dilihat bahwa:
*   Mayoritas responden berusia antara 40 dan 60 tahun.
*   Persentase responden usia produktif (15-64 tahun) lebih tinggi daripada persentase responden usia non-produktif (0-14 tahun dan 65 tahun ke atas).

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Smoking', data=data, palette='pastel')
plt.title('Distribusi Kebiasaan Merokok')
plt.xlabel('Kebiasaan Merokok')
plt.ylabel('Jumlah Responden')
plt.show()
```

Grafik batang tersebut menunjukkan distribusi kebiasaan merokok berdasarkan jenis kelamin. Grafik tersebut menunjukkan bahwa mayoritas dari responden tidak pernah memiliki riwayat merokok.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Glucose', data=data, color='lightgreen')
plt.title('Sebaran Tingkat Glukosa Darah')
plt.xlabel('Tingkat Glukosa Darah')
plt.show()
```

Hasil boxplot di atas menunjukkan sebaran tingkat glukosa darah pada dua kelompok orang, yaitu kelompok yang mengonsumsi lebih banyak gula dan kelompok yang mengonsumsi lebih sedikit gula.

Dari boxplot tersebut, dapat dilihat bahwa:
*   Rata-rata tingkat glukosa darah kelompok yang mengonsumsi lebih banyak gula lebih tinggi daripada kelompok yang mengonsumsi lebih sedikit gula.
*   Distribusi tingkat glukosa darah kelompok yang mengonsumsi lebih banyak gula lebih menyebar daripada kelompok yang mengonsumsi lebih sedikit gula.
*   Adanya outlier pada kelompok yang mengonsumsi lebih banyak gula, yaitu data tingkat glukosa darah yang berada di luar kotak.

Berdasarkan hasil tersebut, dapat disimpulkan bahwa konsumsi gula yang berlebihan dapat meningkatkan tingkat glukosa darah. Hal ini dapat meningkatkan risiko terjadinya penyakit diabetes.

```python
sns.scatterplot(data=df, x='BMI', y='HbA1c', hue='Diabetes')
plt.title('Hubungan BMI dan Kadar HbA1c')
plt.xlabel('BMI')
plt.ylabel('Kadar HbA1c')
plt.show()
```

Scatter plot di atas menunjukkan hubungan antara tingkat glukosa darah dan berat badan pada penderita diabetes.

Dari scatter plot tersebut, dapat dilihat bahwa:
*   Terdapat hubungan positif antara tingkat HbA1c dan berat badan.
*   Semakin berat badan seseorang, semakin tinggi tingkat HbA1c.

Berdasarkan hasil tersebut, dapat disimpulkan bahwa obesitas merupakan salah satu faktor risiko diabetes.

```python
plt.figure(figsize=(8, 5))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap Korelasi')
plt.show()
```


Heatmap korelasi di atas menunjukkan hubungan antara berbagai faktor risiko diabetes, yaitu usia, hipertensi, penyakit jantung, indeks massa tubuh (BMI), hemoglobin A1c (HbA1c), dan kadar glukosa darah.

Dari heatmap tersebut, dapat dilihat bahwa:
*   Terdapat hubungan positif antara semua faktor risiko tersebut.
*   Semakin tinggi nilai suatu faktor risiko, semakin tinggi kemungkinan seseorang menderita diabetes.

Berdasarkan hasil tersebut, dapat disimpulkan bahwa semua faktor risiko tersebut dapat meningkatkan risiko diabetes.

```python
sns.countplot(data=df, x='diabetes', hue='hypertension')
plt.title('Hubungan Diabetes dan Hipertensi')
plt.xlabel('Diabetes')
plt.ylabel('Jumlah')
plt.show()
```

Grafik di atas menunjukkan hubungan antara diabetes dan hipertensi. Grafik tersebut menunjukkan bahwa:
*   Jumlah orang dengan diabetes yang juga menderita hipertensi lebih tinggi daripada jumlah orang dengan diabetes yang tidak menderita hipertensi.
*   Berdasarkan hasil tersebut, dapat disimpulkan bahwa hipertensi merupakan faktor risiko diabetes.
